## IFRS9 macro-economic forecast

### Version Control

#### Introduction

PD's in IFRS9 are so-called point in time versus the long range estimates (through the cycle PD's) that are used in the context of Basel. One particular approach to move from TTC to PIT is to model the dependency of the PD's with respect to macro economic variables and then forecast the macro-economic variables in time. 

```python
cpi_ds =  pd.read_csv('C:\\Users\\nassi\\import\\YALE-CPIQ.csv', parse_dates=['Date']).sort_values('Date', ascending=True)
```

```python
cpi_ds.shape
```

```python
cpi_ds['Days'] = 0
for i in range(0, len(cpi_ds)-1) :
    cpi_ds['Days'][i] = (cpi_ds['Date'][i] - cpi_ds['Date'][i+1]).days
```

```python
cpi_ds['CPI 1M'] = cpi_ds['CPI']
cpi_ds['CPI 2M'] = cpi_ds['CPI']
cpi_ds['CPI 3M'] = cpi_ds['CPI']
cpi_ds['CPI 4M'] = cpi_ds['CPI']
cpi_ds['CPI 5M'] = cpi_ds['CPI']
cpi_ds['CPI 6M'] = cpi_ds['CPI']
cpi_ds['CPI 7M'] = cpi_ds['CPI']
cpi_ds['CPI 8M'] = cpi_ds['CPI']
cpi_ds['CPI 9M'] = cpi_ds['CPI']
cpi_ds['CPI 10M'] = cpi_ds['CPI']
cpi_ds['CPI 11M'] = cpi_ds['CPI']
cpi_ds['CPI 12M'] = cpi_ds['CPI']

for j in range(0, len(cpi_ds) -1) :
    cpi_ds['CPI 1M'][j] = cpi_ds['CPI'][j] + (30 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 2M'][j] = cpi_ds['CPI'][j] + (60 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 3M'][j] = cpi_ds['CPI'][j] + (90 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 4M'][j] = cpi_ds['CPI'][j] + (120 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 5M'][j] = cpi_ds['CPI'][j] + (150 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 6M'][j] = cpi_ds['CPI'][j] + (180 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 7M'][j] = cpi_ds['CPI'][j] + (210 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 8M'][j] = cpi_ds['CPI'][j] + (240 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 9M'][j] = cpi_ds['CPI'][j] + (270 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 10M'][j] = cpi_ds['CPI'][j] + (300 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 11M'][j] = cpi_ds['CPI'][j] + (330 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
    cpi_ds['CPI 12M'][j] = cpi_ds['CPI'][j] + (360 * (cpi_ds['CPI'][j] - cpi_ds['CPI'][j+1]))/cpi_ds['Days'][j]
```
